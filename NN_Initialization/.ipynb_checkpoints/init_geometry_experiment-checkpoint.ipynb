# NN Initialization & Data Geometry Experiment

## Research Question

Does the choice of weight initialization scheme destroy the geometric separability of class data **before any training begins**? Two classes that are linearly separable in input space may become entangled after a single forward pass through a randomly initialized MLP — or they may remain well-separated depending on how the weights were set. This notebook measures class separability (via silhouette score) at every layer of the network for five different initialization strategies, with no gradient updates applied.

## Hypothesis

- **He-normal** and **orthogonal** inits will best preserve separability through deeper layers because they maintain activation variance.
- **Random-normal (small std)** will progressively collapse embeddings toward zero, destroying separability.
- **Zeros** init is the pathological baseline: all activations are identical regardless of input, so silhouette score is undefined (or −1).
- **Xavier** will perform well on early layers but may degrade slightly because it targets tanh/sigmoid, not ReLU.

```python
import os
import random
import warnings

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_classification, fetch_openml
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

warnings.filterwarnings('ignore')
matplotlib.rcParams['figure.dpi'] = 100
os.makedirs('./outputs', exist_ok=True)

print('✓ All imports successful')
```

```python
CONFIG = {
    'seed': 42,
    'n_samples': 1000,
    'n_features': 20,
    'n_informative': 2,
    'hidden_dim': 128,
    'n_layers': 3,
    'batch_size': 256,
    # Toggle 'make_classification' or 'mnist'
    'dataset_name': 'make_classification',
    'init_schemes': [
        'random_normal',
        'xavier_uniform',
        'he_normal',
        'orthogonal',
        'zeros',
    ],
    'pca_components_silhouette': 50,
    'pca_components_viz': 2,
}

print('✓ CONFIG defined')
for k, v in CONFIG.items():
    print(f'  {k}: {v}')
```

```python
def set_all_seeds(seed: int) -> None:
    """Set all random seeds for full reproducibility."""
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

set_all_seeds(CONFIG['seed'])
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'✓ Seeds set to {CONFIG["seed"]}')
print(f'✓ Device: {DEVICE}')
print('✓ Cell group 1 (Setup) complete')
```

## Cell Group 2 — Dataset

We use `sklearn.make_classification` with `n_informative=2` as the primary dataset. This is ideal for the experiment because:

1. The two informative features encode the true class boundary, so a well-behaved init should preserve (or improve) separation of those directions in latent space.
2. The remaining 18 features are noise/redundant, testing whether the network can amplify signal or drowns it.
3. Being synthetic it is fully reproducible and requires no download.

A second option (MNIST digits 0 vs 1) can be enabled via `CONFIG['dataset_name'] = 'mnist'` for a real-world sanity check.


```python
def load_dataset(config: dict):
    """Load and standardise the dataset specified in config."""
    name = config['dataset_name']
    if name == 'make_classification':
        X, y = make_classification(
            n_samples=config['n_samples'],
            n_features=config['n_features'],
            n_informative=config['n_informative'],
            n_redundant=2,
            n_repeated=0,
            n_classes=2,
            random_state=config['seed'],
        )
    elif name == 'mnist':
        mnist = fetch_openml('mnist_784', version=1, as_frame=False, parser='auto')
        mask = (mnist.target == '0') | (mnist.target == '1')
        X = mnist.data[mask][:config['n_samples']].astype(float)
        y = (mnist.target[mask][:config['n_samples']] == '1').astype(int)
    else:
        raise ValueError(f"Unknown dataset: {name}")

    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    return X.astype(np.float32), y.astype(int)


set_all_seeds(CONFIG['seed'])
X_data, y_data = load_dataset(CONFIG)
print(f'✓ Dataset "{CONFIG["dataset_name"]}" loaded')
print(f'  Shape: {X_data.shape}')
print(f'  Class distribution: {dict(zip(*np.unique(y_data, return_counts=True)))}')
```

```python
def plot_raw_data(X: np.ndarray, y: np.ndarray, save_path: str) -> np.ndarray:
    """PCA-project X to 2D and scatter-plot by class. Returns the 2D coords."""
    pca = PCA(n_components=2, random_state=CONFIG['seed'])
    X_2d = pca.fit_transform(X)

    fig, ax = plt.subplots(figsize=(7, 5))
    for cls, colour, label in zip([0, 1], ['#E74C3C', '#2980B9'], ['Class 0', 'Class 1']):
        mask = y == cls
        ax.scatter(X_2d[mask, 0], X_2d[mask, 1],
                   c=colour, alpha=0.65, s=25, label=label)
    ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} var)')
    ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} var)')
    ax.set_title('Raw Input Data — PCA 2D Projection')
    ax.legend()
    ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150, bbox_inches='tight')
    plt.show()
    print(f'  Saved → {save_path}')
    return X_2d


X_2d_raw = plot_raw_data(X_data, y_data, './outputs/00_raw_data.png')
```

```python
score_baseline = silhouette_score(X_2d_raw, y_data)
print(f'Baseline silhouette score (raw 2D PCA of input): {score_baseline:.4f}')
print('  Interpretation: 1.0 = perfect separation | 0.0 = touching | -1.0 = swapped')
print('✓ Cell group 2 (Dataset) complete')
```

## Cell Group 3 — Model Architecture

We use a **minimal 3-layer MLP with ReLU activations** and no output head. Keeping the network small and shallow serves the experiment:

- Deep enough to see geometry evolve across multiple layers.
- Shallow enough to avoid vanishing/exploding gradient effects that would confound the init signal.
- No batch-norm, dropout, or skip connections — we want the raw effect of init on geometry, undiluted by stabilising components.

The critical method is `get_layer_embeddings(x)`, which intercepts activations **after each linear transform but before ReLU**, giving us the full geometry including negative half-space.


```python
class MLP(nn.Module):
    """Simple MLP for init geometry analysis.

    Parameters
    ----------
    input_dim : int
    hidden_dim : int
    n_layers : int  Number of hidden linear layers.
    """

    def __init__(self, input_dim: int, hidden_dim: int, n_layers: int) -> None:
        super().__init__()
        dims = [input_dim] + [hidden_dim] * n_layers
        self.layers = nn.ModuleList(
            [nn.Linear(dims[i], dims[i + 1]) for i in range(len(dims) - 1)]
        )
        self.relu = nn.ReLU()

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        for i, layer in enumerate(self.layers):
            x = layer(x)
            if i < len(self.layers) - 1:
                x = self.relu(x)
        return x

    def get_layer_embeddings(self, x: torch.Tensor) -> dict:
        """Return pre-activation embeddings keyed by layer name.

        Returns a dict with keys 'input', 'layer_1', 'layer_2', ...
        Values are tensors cloned from the computation graph.
        """
        embs = {'input': x.clone()}
        for i, layer in enumerate(self.layers):
            x = layer(x)
            embs[f'layer_{i + 1}'] = x.clone()
            if i < len(self.layers) - 1:
                x = self.relu(x)
        return embs


# Quick sanity check
_test_model = MLP(CONFIG['n_features'], CONFIG['hidden_dim'], CONFIG['n_layers'])
_test_x = torch.randn(10, CONFIG['n_features'])
_test_embs = _test_model.get_layer_embeddings(_test_x)
print('✓ MLP defined and sanity-checked')
print(f'  Layer names: {list(_test_embs.keys())}')
print(f'  Shapes: {[v.shape for v in _test_embs.values()]}')
del _test_model, _test_x, _test_embs
print('✓ Cell group 3 (Model) complete')
```

## Cell Group 4 — Initialization Schemes

| Scheme | Description |
|---|---|
| `random_normal` | **N(0, 0.01)** — naïvely small weights. Activations shrink exponentially with depth; gradients vanish easily. |
| `xavier_uniform` | **Uniform[−k, k], k=√(6/(fan_in+fan_out))** — designed to keep variance stable for tanh/sigmoid; slightly sub-optimal for ReLU. |
| `he_normal` | **N(0, √(2/fan_in))** — the ReLU-aware variant; the factor of 2 compensates for ReLU zeroing the negative half. |
| `orthogonal` | Weight matrix is an **orthogonal matrix** (via QR decomposition). Preserves input norms exactly; the gold standard for preventing collapse. |
| `zeros` | **All weights = 0** — the pathological baseline. Every neuron outputs the same value regardless of input; class signal is completely erased. |


```python
def apply_init(model_net: nn.Module, scheme_name: str) -> nn.Module:
    """Apply named init scheme to all Linear layers in model_net in-place."""
    for module in model_net.modules():
        if not isinstance(module, nn.Linear):
            continue
        if scheme_name == 'random_normal':
            nn.init.normal_(module.weight, mean=0.0, std=0.01)
        elif scheme_name == 'xavier_uniform':
            nn.init.xavier_uniform_(module.weight)
        elif scheme_name == 'he_normal':
            nn.init.kaiming_normal_(module.weight, mode='fan_in', nonlinearity='relu')
        elif scheme_name == 'orthogonal':
            nn.init.orthogonal_(module.weight)
        elif scheme_name == 'zeros':
            nn.init.zeros_(module.weight)
        else:
            raise ValueError(f'Unknown init scheme: {scheme_name}')
        nn.init.zeros_(module.bias)
    return model_net


print('✓ apply_init() defined')
print(f'  Supported schemes: {CONFIG["init_schemes"]}')
print('✓ Cell group 4 (Init schemes) complete')
```

## Cell Group 5 — The Experiment Loop

### What we measure

**Silhouette score** is the right metric here because:

- It measures *relative* inter-class vs intra-class distance without assuming linearity — useful when embeddings may be curved/folded.
- It is defined on any metric space; we use Euclidean distance after PCA-reducing to ≤50 dimensions (to avoid the curse of dimensionality).
- Range [−1, 1]: positive = classes are separated, 0 = overlapping, negative = misclassified clusters.

### Procedure (per init scheme)

1. Instantiate a fresh MLP and apply the init scheme.
2. Run a single forward pass over the **entire dataset** with `torch.no_grad()`.
3. Intercept embeddings at each layer via `get_layer_embeddings()`.
4. Reduce to ≤50 PCA dimensions (silhouette is expensive in high-d).
5. Compute silhouette score. If all embeddings collapsed to a constant (zeros init), record NaN.
6. Also compute a 2D PCA projection for the scatter-plot grid.


```python
layer_names = ['input'] + [f'layer_{i + 1}' for i in range(CONFIG['n_layers'])]
X_tensor = torch.FloatTensor(X_data).to(DEVICE)

# results[scheme][layer] = silhouette score (float or nan)
results: dict = {}
# emb_2d[scheme][layer] = (n_samples, 2) numpy array for scatter plots
emb_2d: dict = {}
# emb_raw[scheme][layer] = full embedding numpy array (needed for norm plot)
emb_raw: dict = {}

for scheme in CONFIG['init_schemes']:
    set_all_seeds(CONFIG['seed'])
    model_scheme = MLP(
        input_dim=X_data.shape[1],
        hidden_dim=CONFIG['hidden_dim'],
        n_layers=CONFIG['n_layers'],
    ).to(DEVICE)
    apply_init(model_scheme, scheme)

    model_scheme.eval()
    with torch.no_grad():
        layer_embs = model_scheme.get_layer_embeddings(X_tensor)

    results[scheme] = {}
    emb_2d[scheme] = {}
    emb_raw[scheme] = {}

    for lname in layer_names:
        emb = layer_embs[lname].cpu().numpy()
        emb_raw[scheme][lname] = emb

        # Check for degenerate (all-same) embeddings
        if np.std(emb) < 1e-8:
            results[scheme][lname] = float('nan')
            emb_2d[scheme][lname] = np.zeros((emb.shape[0], 2))
            continue

        # PCA to ≤50d for silhouette
        n_comp = min(CONFIG['pca_components_silhouette'], emb.shape[1], emb.shape[0] - 1)
        try:
            if emb.shape[1] > n_comp:
                emb_reduced = PCA(n_components=n_comp,
                                  random_state=CONFIG['seed']).fit_transform(emb)
            else:
                emb_reduced = emb
            score_val = silhouette_score(emb_reduced, y_data)
            results[scheme][lname] = float(score_val)
        except Exception:
            results[scheme][lname] = float('nan')

        # 2D PCA for visualisation
        try:
            emb_2d[scheme][lname] = PCA(
                n_components=2, random_state=CONFIG['seed']
            ).fit_transform(emb)
        except Exception:
            emb_2d[scheme][lname] = np.zeros((emb.shape[0], 2))

    summary_row = {k: (f'{v:.3f}' if not np.isnan(v) else 'NaN')
                   for k, v in results[scheme].items()}
    print(f'  {scheme}: {summary_row}')

print('✓ Cell group 5 (Experiment loop) complete')
```

## Cell Group 6 — Visualizations

Four plots, each saved to `./outputs/`:

1. **`01_silhouette_by_layer.png`** — main result: silhouette score vs layer depth per scheme.
2. **`02_embeddings_grid.png`** — PCA scatter grid; rows = schemes, columns = layers.
3. **`03_class_overlap_heatmap.png`** — colour-coded summary table, thesis-ready.
4. **`04_zeros_collapse.png`** — embedding norms for zeros init; shows the pathological collapse.


```python
SCHEME_COLORS = {
    'random_normal':  '#E74C3C',
    'xavier_uniform': '#F39C12',
    'he_normal':      '#27AE60',
    'orthogonal':     '#2980B9',
    'zeros':          '#8E44AD',
}

fig, ax = plt.subplots(figsize=(10, 6))

x_ticks = range(len(layer_names))
for scheme in CONFIG['init_schemes']:
    scores = [results[scheme][l] for l in layer_names]
    ax.plot(x_ticks, scores, marker='o', linewidth=2.5, markersize=7,
            label=scheme, color=SCHEME_COLORS[scheme])

ax.axhline(y=score_baseline, color='black', linestyle='--', linewidth=1.5,
           alpha=0.7, label=f'Baseline (raw input, {score_baseline:.3f})')

ax.set_xticks(x_ticks)
ax.set_xticklabels(layer_names, rotation=30, ha='right')
ax.set_xlabel('Layer', fontsize=12)
ax.set_ylabel('Silhouette Score', fontsize=12)
ax.set_title('Class Separability Across Layers by Initialization Scheme', fontsize=13)
ax.set_ylim(-1.05, 1.05)
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=10)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('./outputs/01_silhouette_by_layer.png', dpi=150, bbox_inches='tight')
plt.show()
print('✓ Saved ./outputs/01_silhouette_by_layer.png')
```

```python
n_schemes = len(CONFIG['init_schemes'])
n_cols = len(layer_names)

fig, axes = plt.subplots(n_schemes, n_cols,
                         figsize=(3.2 * n_cols, 2.8 * n_schemes),
                         squeeze=False)

CLASS_COLORS = np.array(['#E74C3C', '#2980B9'])

for row, scheme in enumerate(CONFIG['init_schemes']):
    for col, lname in enumerate(layer_names):
        ax = axes[row][col]
        proj = emb_2d[scheme][lname]
        for cls in [0, 1]:
            mask = y_data == cls
            ax.scatter(proj[mask, 0], proj[mask, 1],
                       c=CLASS_COLORS[cls], s=8, alpha=0.55)
        ax.set_xticks([])
        ax.set_yticks([])
        score_txt = results[scheme][lname]
        score_str = f'{score_txt:.2f}' if not np.isnan(score_txt) else 'NaN'
        ax.set_title(f'{lname}\ns={score_str}', fontsize=7.5)
        if col == 0:
            ax.set_ylabel(scheme, fontsize=8, rotation=45, ha='right', labelpad=40)

fig.suptitle('Embedding Geometry: PCA 2D Projections\n'
             '(Rows = Init Schemes, Columns = Layers)', fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('./outputs/02_embeddings_grid.png', dpi=150, bbox_inches='tight')
plt.show()
print('✓ Saved ./outputs/02_embeddings_grid.png')
```

```python
scores_matrix = pd.DataFrame(
    {scheme: [results[scheme][l] for l in layer_names]
     for scheme in CONFIG['init_schemes']},
    index=layer_names,
).T

fig, ax = plt.subplots(figsize=(9, 4))
sns.heatmap(
    scores_matrix.astype(float),
    annot=True,
    fmt='.3f',
    cmap='RdYlGn',
    vmin=-1, vmax=1,
    linewidths=0.5,
    ax=ax,
    cbar_kws={'label': 'Silhouette Score', 'shrink': 0.8},
)
ax.set_title('Class Overlap Heatmap — Silhouette Score by Init Scheme and Layer',
             fontsize=12)
ax.set_xlabel('Layer', fontsize=11)
ax.set_ylabel('Init Scheme', fontsize=11)
plt.tight_layout()
plt.savefig('./outputs/03_class_overlap_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print('✓ Saved ./outputs/03_class_overlap_heatmap.png')
```

```python
zeros_norms = []
for lname in layer_names:
    emb = emb_raw['zeros'][lname]
    norms = np.linalg.norm(emb, axis=1)
    zeros_norms.append(norms.mean())

# Compare against other schemes for context
fig, axes_z = plt.subplots(1, 2, figsize=(13, 5))

# Left: zeros collapse
ax_left = axes_z[0]
ax_left.plot(range(len(layer_names)), zeros_norms,
             marker='o', color='#8E44AD', linewidth=2.5, markersize=9)
ax_left.fill_between(range(len(layer_names)), zeros_norms, alpha=0.15, color='#8E44AD')
ax_left.set_xticks(range(len(layer_names)))
ax_left.set_xticklabels(layer_names, rotation=30, ha='right')
ax_left.set_xlabel('Layer', fontsize=11)
ax_left.set_ylabel('Mean Embedding L2-Norm', fontsize=11)
ax_left.set_title('Zeros Init: Norm Collapse Across Layers\n(Pathological Baseline)', fontsize=12)
ax_left.grid(alpha=0.3)

# Right: all schemes for comparison
ax_right = axes_z[1]
for scheme in CONFIG['init_schemes']:
    norms_per_layer = [
        np.linalg.norm(emb_raw[scheme][l], axis=1).mean()
        for l in layer_names
    ]
    ax_right.plot(range(len(layer_names)), norms_per_layer,
                  marker='o', linewidth=2, label=scheme,
                  color=SCHEME_COLORS[scheme])
ax_right.set_xticks(range(len(layer_names)))
ax_right.set_xticklabels(layer_names, rotation=30, ha='right')
ax_right.set_xlabel('Layer', fontsize=11)
ax_right.set_ylabel('Mean Embedding L2-Norm', fontsize=11)
ax_right.set_title('Embedding Norms: All Init Schemes', fontsize=12)
ax_right.legend(fontsize=9)
ax_right.grid(alpha=0.3)

plt.suptitle('Embedding Norm Analysis — Zeros Init Collapse vs Others', fontsize=13)
plt.tight_layout()
plt.savefig('./outputs/04_zeros_collapse.png', dpi=150, bbox_inches='tight')
plt.show()
print('✓ Saved ./outputs/04_zeros_collapse.png')
print('✓ Cell group 6 (Visualizations) complete')
```

## Cell Group 8 — Next Steps

Three natural extensions of this experiment for the thesis:

### 1. Repeat with deeper networks (10, 20, 50 layers)
The 3-layer setting here shows the effect clearly but may understate it. Deeper networks amplify the difference between inits dramatically: random-normal and zeros will collapse to zero much faster, while orthogonal should maintain norms for many more layers. This tests the scalability of the finding.

### 2. Repeat with different activation functions (sigmoid, tanh, GELU)
ReLU was used here because He-normal is designed for it. Swapping to tanh should make Xavier the best performer, and GELU (used in transformers) may behave differently from both. Comparing activations directly shows the *interaction* between activation choice and init choice.

### 3. Compare with a trained network — does training recover geometry lost by init?
The fundamental thesis question: if zeros init collapses the geometry completely, can gradient descent recover it with enough epochs? Measuring silhouette score at checkpoints (epoch 0, 1, 5, 20, 100) would reveal whether bad init permanently limits what the network can represent, or whether it is merely a slower start.
